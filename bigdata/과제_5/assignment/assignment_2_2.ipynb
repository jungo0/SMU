# 3. 전체 학생들의 math, reading, writing 점수의 평균 구하기

```python
avg_math = student_df['math score'].mean()
avg_reading = student_df['reading score'].mean()
avg_writing = student_df['writing score'].mean()

print('전체 학생들의 math 점수 평균 :', avg_math)
print('전체 학생들의 reading 점수 평균 :', avg_reading)
print('전체 학생들의 writing 점수 평균 :', avg_writing)
```

# 6. 남자들 중에서 수학 점수가 80점 이상인 학생들은 몇 명인가요?

```python
male_student_math_gt_80 = student_df[(student_df['gender'] == 'male') & (student_df['math score'] >= 80)].count()

male_student_math_gt_80
```

```python
male_student_math_gt_80.gender
```
