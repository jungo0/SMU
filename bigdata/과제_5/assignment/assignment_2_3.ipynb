```python
import numpy as np
```

# 1. 정규분포를 따르는 임의의 값을 가지는 행렬 x1 = (100,10)을 만드세요.

```python
x1 = np.random.rand(100,10)

print(x1.shape)
print(x1)
```

# 2. 행렬 x1에서 행(20-70번째), 열(1-6) 까지만 잘라내세요.

```python
x1 = x1[20:70, 1:6]

print(x1.shape)
print(x1)
```

# 3. 잘라낸 행렬과 내적곱을 계산할 행렬 x2 = (?, 5)를 만드세요.

### x2의 값은 내적곱을 계산할 수 있는 알맞은 값을 찾으세요

```python
#(50, 5) * (?, 5)
#(50, 5) * (5, 5) = (50, 5)

x2 = np.random.randn(5,5)

x2
```

# 4. x1과 x2을 서로 내적곱 y를 구하세요. 

```python
y = np.dot(x1, x2)

print(y.shape)
print(y)
```

# 5. y의 전치행렬을 구하세요

```python
yT = (y.T)

print(yT.shape)
print(yT)
```
