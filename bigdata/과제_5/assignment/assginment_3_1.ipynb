```python
from selenium import webdriver
import os
import time

searchTerm = "음식"

url = "https://search.naver.com/search.naver?where=image&sm=tab_jum&query=%EC%9D%8C%EC%8B%9D"

browser = webdriver.Chrome(executable_path='/Users/ajy/bigdata/chromedriver')
browser.get(url)

for _ in range(2):
    browser.execute_script("window.scrollTo(0, document.body.scrollHeight);")
    time.sleep(5)

#for _ in range(5):
    #browser.execute_script("window.scrollTo(0, window.scrollY + 1000);")
    #time.sleep(3)
    
if not os.path.exists(searchTerm):
    os.mkdir(searchTerm)

succounter = 0

#body = browser.find_element_by_id('container')
#img_list = body.find_elements_by_tag_name('img')
img_list = browser.find_elements_by_tag_name('img')

for idx, value in enumerate(img_list):
    img_src = value.get_attribute('src')
    print(img_src)
    try:
        save_path = os.path.join(searchTerm, str(idx) +".png")
        print(save_path)
        
        value.screenshot(save_path)
        time.sleep(2)
        succounter += 1

        running_test = "running: " + str(succounter) + " /" + str(len(img_list))
        print(running_test)

    except:
        print("can't get img")

browser.close()
```
