```python
malignant = breast_cancer.data[breast_cancer.target == 0]
benign = breast_cancer.data[breast_cancer.target == 1]

print(malignant.shape)
print(benign.shape)
```

```python
df = pd.DataFrame(data = breast_cancer.data,
                 columns = breast_cancer.feature_names)
```

```python
print (df)
```

```python
import matplotlib.pyplot as plt
import numpy as np

plt.figure(figsize=[20,40])  #가로 20 세로 40

for col in range(30):    #feature 값이 30
    plt.subplot(10,3,col+1)   #행 10개 열 3개
    _, bins=np.histogram(breast_cancer.data[:,col],bins=20)
    
    plt.hist(malignant[:,col],bins=bins, alpha=0.3)
    plt.hist(benign[:,col],bins=bins, alpha=0.3)
    plt.title(breast_cancer.feature_names[col])
if col==0: 
    plt.legend(breast_cancer.target_names) 
    plt.xticks([])
```

```python
correlation = np.corrcoef(breast_cancer.data.T)
print(correlation)
```

```python
fig =plt.figure(figsize=[14,14])
plt.title('Breast Cancer - Correlation Coefficient',fontsize=20)
plt.imshow(correlation, interpolation='none', vmin=-1, vmax=1)
plt.colorbar(shrink=0.7)
plt.xticks(range(30), breast_cancer.feature_names,rotation=90,ha='center')
plt.yticks(range(30), breast_cancer.feature_names)
print('')
```

```python
from sklearn.model_selection import train_test_split
## Train : Test = 80:20  학습이랑 테스트값에 데이터값을 다르게해야댐
x_train, x_test, y_train, y_test = train_test_split(
    breast_cancer.data, breast_cancer.target, test_size = 0.2)   #class값 feature값
```

```python
print (len(x_train), len(x_test))
```

```python
from sklearn.linear_model import LogisticRegression 
model = LogisticRegression()   #학슴댐
model.fit(x_train, y_train)
```

```python
##정확도 , test set 사용
model.score(x_test,y_test)
```

```python
pd.Series(model.coef_[0], breast_cancer.feature_names).sort_values(ascending=False)     
#30개의 feature coef값을 살펴봄 , 더 영향을 주는순으로
```
