```python
import os
import time
from selenium import webdriver
```

```python
'''
크롬드라이버 다운로드
1. 크롬을 최신화 해준다
2. selenium 홈페이지 접속
3. Selenium WebDriver
4. Browsers -> Chrome / https://sites.google.com/a/chromium.org/chromedriver
   맥의 경우도 동일(경로 설정은 http://dbjina.tistory.com/50 참고. 본인이 직접 아무 경로로 설정해도 무관)
'''

#찾고자 하는 검색얼ㄹ url로 만들어 준다.
searchterm = "강아지"

#URL 작성
#searchterm은 parameter 중의 하나, url의 경우 & 기호로 parameter를 구분.
url = "https://www.google.com/search?q="+searchterm+"&source=lnms&tbm=isch"
```

```python
#Chrome Webdriver 실행
browser = webdriver.Chrome(executable_path="chromedriver.exe")
browser.get(url)
```

```python
#검색 결과를 늘리기 위하여 스크롤 다운
for _ in range(2):
    browser.execute_script("window.scrollTo(0, document.body.scrollHeight);")
    time.sleep(2)
```

```python
# 소스코드가 있는 경로에 '검색어' 폴더가 없으면 만들어준다.(크롤링한 이미지 저장을 위해서)
if not os.path.exists(searchterm):
    os.mkdir(searchterm)
```

```python
#현재 수집된 이미지 수를 나타내는 변수(진행상황 체크를 위해 필요. 없어도 상관은 없음.)
succounter = 0

#이미지 크롤링(body를 전체로 놓고 하면 구글 로고나 썸네일 같은 이미지도 모두 포함되므로 필요한 부분만)
body = browser.find_element_by_id('islrg')
img_list = body.find_elements_by_tag_name('img')


for idx, value in enumerate(img_list):
    img_src = value.get_attribute('src')
    print(img_src)
    try:
        save_path = os.path.join(searchterm, str(idx)+".png")
        print(save_path)

        #이미지를 저장
        value.screenshot(save_path)
        succounter = succounter + 1

        running_test = "running: " + str(succounter) +" / " + str(len(img_list))
        print(running_test)

    except:
        print ("can't get img")
            

print (succounter, "succesfully downloaded")

# 크롬 브라우저 종료
browser.close()
```
