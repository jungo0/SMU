```python
import os
import time
from selenium import webdriver
```

```python
searchterm='음식'
url='https://search.naver.com/search.naver?where=image&sm=tab_jum&query='+searchterm+'&source=lnms&tbm=isch'
```

```python
browser=webdriver.Chrome(executable_path='chromedriver.exe')
browser.get(url)
```

```python
for _ in range(2):
    browser.execute_script("window.scrollTo(0, document.body.scrollHeight);") #window 창의 0부터 ㅊ
    time.sleep(2)
```

```python
if not os.path.exists(searchterm):
    os.mkdir(searchterm)
```

```python
succounter=0

body=browser.find_element_by_id('main_pack')
img_list=body.find_elements_by_tag_name('img')

for idx,value in enumerate(img_list):
    img_src=value.get_attribute('src')
    print(img_src)
    try:
        save_path=os.path.join(searchterm,str(idx)+'.png')
        print(save_path)

        value.screenshot(save_path)
        succounter=succounter+1

        runing_test='running'+str(succounter)+'/'+str(len(img_list))
        print(running_test)

    except:
        print('cant get img')
```
